# Árvore

Uma **árvore** é uma estrutura de dados hierárquica e acíclica composta por **nós**, e existem alguns tipos especiais de nós como o nó **raíz** e o nó **folha**.

```python
from anytree import Node, RenderTree
%matplotlib inline
```

```python
r = Node('Nó Raiz')
n1 = Node('Nó A', parent=r)
n2 = Node('Nó B', parent=r)
n3 = Node('Nó C', parent=r)

n11 = Node('Nó folha 1', parent=n1)
n12 = Node('Nó folha 2', parent=n1)

n21 = Node('Nó folha 3', parent=n2)

n31 = Node('Nó D', parent=n3)
n311 = Node('Nó folha 4', parent=n31)
```

```python
for pre, fill, node in RenderTree(r):
    print("%s%s" % (pre, node.name))
```

```python
from anytree.exporter import DotExporter
k = DotExporter(r)
```

```python
k.to_dotfile('tree.dot')
```

```python
import graphviz
```

```python
f = open("tree.dot")
dot_graph = f.read()
f.close()

graphviz.Source(dot_graph)
```

Cada nó possui um nó **pai** (com exceção do nó raiz) e pode possuir diversos nós **filhos**. O Nó raiz é o nó inicial de uma árvore, e os nós folha são os nós finais.

Chama-se de **profundidade** da árvore o número de nós presentes no caminho com maior número de nós do nó raíz até o nó folha folha.

**Exemplo**  


```python
print(n11)
print(n11.depth)
print(n12)
print(n21)
print(n311)
```

```python
print(r.descendants)
```

```python
for folha in r.leaves:
    print(f'Caminho:{folha} - profundidade: {folha.depth}')
```

```python
for no in r.children:
    print(f'Caminho:{no} - filhos: {no.height}')
```
