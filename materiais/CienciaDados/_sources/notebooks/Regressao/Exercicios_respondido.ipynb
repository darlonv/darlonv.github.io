```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Regressão linear
from sklearn import metrics #Cálculo do erro
```

Utilizando regressão linear para estimar valores, responda as questões apresentadas abaixo. Utilize a base de dados `rioAptos`, disponibilizada em [BasesPublicas](https://drive.google.com/drive/folders/166mbePwjfMhmCOR0ogv2n9rNLtv3uiJB).

- Qual é o valor estimado do condomínio de um apartamento que custa 90 mil reais no bairro da Tijuca?
    - Apresente o gráfico de dispersão dos dados, em conjunto com:
        - a reta que representa a regressão linear entre as duas variáveis consideradas
        - um ponto destacando o valor estimado para o valor de apartamento de 90000 reais
    - Qual é o erro médio absoluto da regressão?

```python
#Carregando base
base = pd.read_csv('./rioAptos.csv')

base= base[ base['bairro'] == 'Tijuca' ]

#Alocando regressão linear
rl = LinearRegression()

#Criando regressão linear
preco = base['preco'].values.reshape(-1,1)
cond  = base['condominio'].values.flatten()
rl.fit(preco, cond)

#Observando novos valores
x = [90000] 
y_pred  = rl.predict(np.array(x).reshape(-1,1)) #calcula a predição
print('Preco : ', x)
print('Cond  : ', y_pred)
```

```python
base.plot.scatter('preco','condominio')

x = [min(preco),max(preco)]
y  = rl.predict(np.array(x).reshape(-1,1)) #calcula a predição
plt.plot(x,y)

x2 = [90000]
y2  = rl.predict(np.array(x2).reshape(-1,1)) #calcula a predição

plt.scatter(x2,y2, color='red')
```

```python
#Calculando o erro médio abosoluto
preco = base['preco'].values.reshape(-1,1)
cond  = base['condominio'].values.flatten()
cond_pred  = rl.predict(np.array(preco).reshape(-1,1)) #calcula a predição

ema = metrics.mean_absolute_error(cond, cond_pred)
print('EMA:', ema)
```

- Se o valor do condomínio de determinado apartamento situado no bairro do Leblon é mil reais, qual seria o valor aproximado do apartamento?

```python
#Carregando base
base = pd.read_csv('/home/darlon/Pesquisa/bases/Real Estate/RioAptos/dados.csv')

base= base[ base['bairro'] == 'Leblon' ]

#Alocando regressão linear
rl = LinearRegression()

#Criando regressão linear
cond  = base['condominio'].values.reshape(-1,1)
preco = base['preco'].values.flatten()

rl.fit(cond, preco)

#Observando novos valores
x = [1000] 
y_pred  = rl.predict(np.array(x).reshape(-1,1)) #calcula a predição
print('Cond  : ', x)
print('Preco : ', y_pred)
```

```python
#Calculando o erro médio abosoluto
cond  = base['condominio'].values.reshape(-1,1)
preco = base['preco'].values.flatten()

preco_pred  = rl.predict(np.array(cond).reshape(-1,1)) #calcula a predição

ema = metrics.mean_absolute_error(preco, preco_pred)
print('EMA:', ema)
```
